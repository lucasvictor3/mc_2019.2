```python
# Biblioteca científica
import numpy as np
# Biblioteca para análise de dados
import pandas as pd
# Suprimindo os warnings da biblioteca pandas
pd.options.mode.chained_assignment = None
# Bibliotecas de visualização de dados
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.ticker as mtick
# import missingno as msno
plt.style.use('ggplot')
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
# Custo das universidades federais do nordeste
custo_ufs = pd.read_csv('../data/custo_ies_federais_nordeste.csv')
# Exibindo o dataframe
custo_ufs
```

```python
# Custo de cada aluno por universidade federal no nordeste
custo_aluno = custo_ufs[['Instituicao', 'custoAluno']]

# Adicionando a coluna de média no enade por curso com valor default zero
custo_aluno['NT_GER'] = 0

# Exibindo o dataframe
custo_aluno
```

```python
# Nota ENADE de cursos de computação nas Universidades Federais do Nordeste
notas_enade_ufs = pd.read_csv('../data/enade2017_ccc_ne_federais_prep.csv')
# Exibindo o dataframe
notas_enade_ufs.head()
```

```python
# Todas as universidades do nordeste
universidades = custo_aluno['Instituicao']

#Exibindo as universidades
universidades
```

```python
# Calculando a média das notas do enade por universidade e atualizando o DataFrame
for uf_nome in universidades:
        # DataFrame de cada universidade com as notas de todos os alunos
        notas_universidade = notas_enade_ufs[notas_enade_ufs['Instituicao'] == uf_nome]
        
        # Atualizando a coluna de média do enade de cada universidade
        custo_aluno.loc[(custo_aluno['Instituicao']) == uf_nome, 'NT_GER'] = notas_universidade[['Instituicao', 'NT_GER']].mean()['NT_GER']
        
# Exibindo dataframe com as médias do enade de cada universidade
custo_aluno
```

```python
# Correlação entre custo do aluno e a média do enade das universidades federais do nordeste
custo_aluno.corr()
```

## Conclusão

Com um correlação de 0.732179 entre as duas variáveis, podemos então afirmar que quanto mais custar um aluno nas universidades federais do nordeste, maior tende a ser a nota do mesmo no enade.
